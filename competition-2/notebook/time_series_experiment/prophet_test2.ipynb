```python
def ts_label(df, year_month, ret_type='bool'):
    df = df.copy()
    df['year_month'] = [x.strftime("%Y-%m") for x in df['order_date'].copy()]
    cust = df[df['year_month']<=year_month]['customer_id'].unique()
    cust.sort()
    df = df[df['year_month']==year_month]
    df = df.groupby("customer_id").sum().reset_index(drop=False)[['customer_id', 'total']]
    
    ret_df = pd.DataFrame({"customer_id":cust})
    ret_df = ret_df.merge(df, on='customer_id', how='left')
    ret_df = ret_df.rename({"total":"label"}, axis='columns')
    ret_df.fillna(0, inplace=True)
    if ret_type=='bool':
        ret_df['label'] = (ret_df>=300).astype(int)
        return ret_df
    elif ret_type=='int':
        return ret_df
    else:
        raise("ret_type error")
        return None
    
```

```python
lower_bd = datetime.date(year=2011, month=12, day=1) - dateutil.relativedelta.relativedelta(days=14)
upper_bd = datetime.date(year=2011, month=12, day=1) - dateutil.relativedelta.relativedelta(days=7)

def ts_df_sampling(df, year_month, period):
    df = df.copy()
    df['y_m_d'] = df['order_date'].dt.strftime("%Y-%m-%d")
    df['y_m'] = df['order_date'].dt.strftime("%Y-%m")
    cust = df[df['y_m']<=year_month]['customer_id'].unique()
    cust.sort()
    ret_dict = {}
    for c_idx in list(cust):
        ret_dict[c_idx] = []
    year, month = map(int, year_month.split('-'))
    ref_date = datetime.date(year=year, month=month, day=1)
    ds_col = []
    for i in range(period[1]):
        lower_bd = ref_date - dateutil.relativedelta.relativedelta(days=(i+1)*period[0])
        upper_bd = ref_date - dateutil.relativedelta.relativedelta(days=(i)*period[0])
        ds_col.append(upper_bd)
        now_df = df[(df.y_m_d<upper_bd.strftime("%Y-%m-%d"))&(df.y_m_d>lower_bd.strftime("%Y-%m-%d"))]
        for c_idx in ret_dict.keys():
            c_sum = now_df[now_df['customer_id']==c_idx]['total'].sum()
            ret_dict[c_idx].append(c_sum)
    cust_col = []
    val_col = []
    for c_idx in list(cust):
        cust_col += [c_idx]*period[1]
        val_col += ret_dict[c_idx]
    ret_df = pd.DataFrame({"customer_id":cust_col, "ds":ds_col*len(cust), "y":val_col})
    return ret_df
```

```python
ret_df = ts_df_sampling(train_csv, "2011-11", (7, 10))
```
